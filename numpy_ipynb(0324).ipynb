## 5. Operations

### 5.1 Arithmetic

```python
a = np.arange(1, 5)         #디폴트는 1이니께
b = np.arange(9, 5, -1)                       #행렬이랑 똑같은가? Vector랑 같은 개념인가???
c = np.arange(1, 5, 0.1)
print(a)
print(b)
print(c)
```

```python

print(a - b)
print(a * b)               # 이것은  행렬의 곱이 아닌 그 안에서의 숫자의 곱! 애네는 매트릭스가 아닌 것이다!
print(a-c)
```

```python
A = np.matrix(a)
B = np.matrix(b)
print(A)
print(B)                    #a는 1*4 b는 1*4
print(B.T)
```

```python
A * B.T                #_____.T를 Transpose를 시키는 것! 1*4에서 4*1로 바꿔주는 것!
```

```python
a.T * b
```
